```python
%matplotlib inline
import pandas as pd
from makerlib import wind_direction_df
from ideas import westerly_count
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] =(12,10)
```

```python
%%time
df = westerly_count(vwnd_file='data/vwnd.IGS.48to17.nc', uwnd_file='data/uwnd.IGS.48to17.nc', 
                  startdate='1948-01-01', enddate='2016-01-01', 
                  level=300, lon=220, lat=50)
```

```python
print '\n Dataframe showing count of days each year/summer/winter where there are westerly (W/WNW/WSW) \n upper atmosphere winds (300hPa) (Cell = 140W 50N)'
df  # Note there is some overlap bewteen year count and DJF count
```

```python
time_series = pd.DataFrame({'summer' : df['JJA_count'], 'winter' : df['DJF_count']})
ax = time_series.plot(color=('y','c'))
plt.xlabel('Year')
plt.ylabel('Count of \'westerly\' days', size=12)
plt.title('Cell = 140W 50N')
# ax.set_facecolor('k')
```

```python
# barplot version
ax = sns.barplot(x="year_count", y=df.index, data=df, orient='h', color='skyblue')
yticks=ax.yaxis.get_major_ticks()
for i in range(len(yticks)):
    if i%2==1:
        yticks[i].set_visible(False)
        
plt.xlabel('Count of \'westerly\' days')
plt.title('Cell = 140W 50N')
plt.show()
```
