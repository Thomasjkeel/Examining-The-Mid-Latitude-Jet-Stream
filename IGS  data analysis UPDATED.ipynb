```python
import pandas as pd 
df = pd.read_pickle('pickle_jars/igs_010148to010117_24050.pkl')
dfW = df.loc[((df['cardinal direction'].str.startswith('W') == True))]

dicts = {}
keys = range(2016-1948) # with winter it becomes difficult 
a = 1948

for i in keys:
    b = '{0}-01-01'.format(str(a))
    c = '{0}-01-01'.format(str(a+1))
    values = len(dfW[b:c])
    dicts[a] = values
    a += 1
    
df_year_counts = pd.DataFrame.from_dict(dicts, orient='index') # for indexing and time series\
df_year_counts.columns = ['year_count']
# df_year_counts = pd.DataFrame(dicts.items(), columns=['year', 'year_count']) # count of W/WSW/WNW
# df_year_counts.year = df_year_counts.astype(str)
```

```python
jja_dicts = {}
a = 1948

for i in keys:
    b = '{0}-06-01'.format(str(a))
    c = '{0}-08-31'.format(str(a))
    values = len(dfW[b:c])
    jja_dicts[a] = values
    a += 1
    
df_summer_counts = pd.DataFrame.from_dict(jja_dicts, orient='index') # for indexing and time series\
df_summer_counts.columns = ['summer_count']   

# df_summer_counts = pd.DataFrame(jja_dicts.items(), columns=['JJA', 'summer_count']) # count of W/WSW/WNW
```

```python
djf_dicts = {}
a = 1948
keys = range(2016-1948)

for i in keys:
    b = '{0}-12-01'.format(str(a))
    c = '{0}-02-28'.format(str(a+1)) # problem that need solving: leap years and 29th February
    values = len(dfW[b:c])
    djf_dicts[a] = values
    a += 1
df_winter_counts = pd.DataFrame.from_dict(djf_dicts, orient='index') # for indexing and time series\
df_winter_counts.columns = ['winter_count']   

# df_winter_counts = pd.DataFrame(djf_dicts.items(), columns=['DJF', 'winter_count']) # count of W/WSW/WNW
```

```python
count_df = df_year_counts.join(df_summer_counts)
count_df = count_df.join(df_winter_counts)
```
